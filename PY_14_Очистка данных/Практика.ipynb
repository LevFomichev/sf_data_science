```python
import pandas as pd

import numpy as np

diabetes = pd.read_csv('data/diabetes_data.csv', sep=',')
```

```python
dupl_diabetes = list(diabetes.columns)
mask = diabetes.duplicated(subset=dupl_diabetes)
diabetes_duplicated = diabetes[mask]
diabetes = diabetes.drop_duplicates(subset=dupl_diabetes)
diabetes.shape
```

```python
#список неинформативных признаков
low_information_cols = [] 

#цикл по всем столбцам
for col in diabetes.columns:
    #наибольшая относительная частота в признаке
    top_freq = diabetes[col].value_counts(normalize=True).max()
    #доля уникальных значений от размера признака
    nunique_ratio = diabetes[col].nunique() / diabetes[col].count()
    # сравниваем наибольшую частоту с порогом
    if top_freq > 0.95:
        low_information_cols.append(col)
    # сравниваем долю уникальных значений с порогом
    if nunique_ratio > 0.95:
        low_information_cols.append(col)
diabetes.drop(low_information_cols, axis=1)
```

```python
def np_nan(arg):
    if arg == 0:
        return np.nan
    else:
        return arg
diabetes['Glucose'] = diabetes['Glucose'].apply(np_nan)
diabetes['BloodPressure'] = diabetes['BloodPressure'].apply(np_nan)
diabetes['SkinThickness'] = diabetes['SkinThickness'].apply(np_nan)
diabetes['Insulin'] = diabetes['Insulin'].apply(np_nan)
diabetes['BMI'] = diabetes['BMI'].apply(np_nan)
insulin_nan = diabetes['Insulin'].isnull().mean()
print(round(insulin_nan, 2))
display(diabetes)
```

```python
thresh = diabetes.shape[0]*0.7
diabetes = diabetes.dropna(thresh=thresh, axis=1)
print(diabetes.shape[1])
```
